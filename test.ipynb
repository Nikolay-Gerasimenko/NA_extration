```python
def get_ld_from_file(document_name):
    return LD(get_text_from_file(document_name))
```

```python
def get_text_from_file(document_name):
    soup = BeautifulSoup(open(document_name), "lxml")
    temp = soup.getText()[soup.getText().find('ВЕРХОВНЫЙ СУД РОССИЙСКОЙ ФЕДЕРАЦИИ'):]
    text = temp[:temp.find('Документ предоставлен КонсультантПлюс')]
    return text
```

```python
df = pd.read_excel('./разметка.xlsx')
```

## Проверка извлечения na_string

```python
i=0
for document_name in document_names:
    fn = []
    tp = []
    fp = []
    
    for na_string in LD.get_na_strings(get_text_from_file(document_name)):
        if na_string not in df[df['document']==document_name]['entity']:
            tp.append(na_string)
        else:
            fp.append(na_string)

    for na_string in df[df['document']==document_name]['entity']:
        if na_string not in LD.get_na_strings(get_text_from_file(document_name)):
            fn.append(na_string)
    
    if i >= 20:#len(fn+fp)!=0: 
        print(i)
        print(document_name)
        print('TP:\n{}'.format(tp))
        print('FP:\n{}'.format(fp))
        print('FN:\n{}'.format(fn))
        print('__________________________________________________')
    
    i+=1
```
